```python
import urllib.request
```

```python
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/SpaceX')
# response has this
# <http.client.HTTPResponse at 0x1a7852f9eb8>
```

```python
html = response.read()
print(html)
```
